```python
data = df_jh.copy()
data.set_index(keys='wqdt',inplace=True)
```

```python
data['bBac'] = data['bBac'].astype('float')
data['bBod'] = data['bBod'].astype('float')
data['temper'] = data['temper'].astype('float')
data['uBod'] = data['uBod'].astype('float')
data['uCod'] = data['uCod'].astype('float')
data['uSs'] = data['uSs'].astype('float')
```

```python
data['uVal'] = data['uVal'].str.replace(',','')
data['uVal'] = data['uVal'].astype('float')
data['uBac'] = data['uBac'].str.replace(',','')
data['uBac'] = data['uBac'].astype('float')
data['bVal'] = data['bVal'].str.replace(',','')
data['bVal'] = data['bVal'].astype('float')
```

```python
data = data.drop(['loccd','locnm'],axis=1)
```

```python
data.weather.astype('category')
col = data.columns.tolist()
data2 = pd.get_dummies(data[col])
```

```python
X = data2.copy()
# X = scaler.fit_transform(X)
# X = pd.DataFrame(X)
# y = X[['bCod']]
# z = X[['bTn']]
# 
```

```python
X = X.drop(['index'],axis=1)
X['weather_눈'] = X['weather_눈'].astype('float')
X['weather_맑음'] = X['weather_맑음'].astype('float')
X['weather_비'] = X['weather_비'].astype('float')
X['weather_흐림'] = X['weather_흐림'].astype('float')

X.head()
```

```python
X.dtypes
```

```python
return_period = 1
y = X.loc[:, ['bCod']].shift(-return_period)
z = X.loc[:,['bTn']].shift(-return_period)

y.columns = [col+'_pred' for col in y.columns]
z.columns = [col+'_pred' for col in z.columns]

#X = X.drop(['bCod','bTn'],axis=1)

dataset = pd.concat([z,y,X],axis=1).dropna().iloc[:,:]

z = dataset.loc[:,z.columns]
y = dataset.loc[:,y.columns]
X = dataset.loc[:,X.columns]
```

```python
X_train = X.iloc[:826]
X_test = X.iloc[826:]
y_train = y.iloc[:826]
y_test = y.iloc[826:]
z_train = z.iloc[:826]
z_test= z.iloc[826:]
```

```python
X_train = np.array(X_train).reshape(len(X_train),1,20)
X_test = np.array(X_test).reshape(len(X_test),1,20)
y_train = np.array(y_train).reshape(len(y_train),1,1)
y_test = np.array(y_test).reshape(len(y_test),1,1)
z_train = np.array(z_train).reshape(len(z_train),1,1)
z_test = np.array(z_test).reshape(len(z_test),1,1)


print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
import tensorflow as tf
from keras.callbacks import EarlyStopping

COD_model = tf.keras.models.Sequential([
    tf.keras.layers.GRU(32,return_sequences=True,activation='relu',input_shape=(1,20)),
    tf.keras.layers.GRU(16,return_sequences=True,activation='relu'),
    tf.keras.layers.GRU(8,return_sequences=False,activation='relu'),
    tf.keras.layers.Dense(1)
])

COD_model.compile(optimizer='adam', loss='mse', metrics=['mae'])
ea = EarlyStopping(monitor='val_loss', patience=50)

COD_history = COD_model.fit(X_train, y_train,
                            validation_data=(X_test,y_test),
                            epochs=100, batch_size=16,
                            callbacks=[ea]
                            )
```

```python
TN_model = tf.keras.models.Sequential([
    tf.keras.layers.GRU(32,return_sequences=True,activation='relu',input_shape=(1,20)),
    tf.keras.layers.GRU(16,return_sequences=True,activation='relu'),
    tf.keras.layers.GRU(8,return_sequences=False,activation='relu'),
    tf.keras.layers.Dense(1)
])

TN_model.compile(optimizer='adam', loss='mse', metrics=['mae'])
ea = EarlyStopping(monitor='val_loss', patience=50)

TN_history = TN_model.fit(X_train, z_train,
                            validation_data=(X_test,z_test),
                            epochs=100, batch_size=16,
                            callbacks=[ea]
                            )
```

```python
plt.figure(figsize=(12,6))

plt.plot(COD_history.history['mae'],color='black',linewidth=2)
plt.plot(COD_history.history['val_mae'],color='blue',linewidth=2)
plt.title('COD_Model History')
plt.xlabel('epoch')
plt.ylabel('MAE')
plt.legend(['train','test'], loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(TN_history.history['mae'],color='black',linewidth=2)
plt.plot(TN_history.history['val_mae'],color='blue',linewidth=2)
plt.title('TN_Model History')
plt.xlabel('epoch')
plt.ylabel('MAE')
plt.legend(['train','test'], loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.title('COD_pred')
plt.ylabel('pred')
plt.plot(y_test.reshape(205,1), label = 'Real')
plt.plot(COD_model.predict(X_test).reshape(205,1), label = 'Predict')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title('TN_pred')
plt.ylabel('pred')
plt.plot(z_test.reshape(205,1), label = 'Real')
plt.plot(TN_model.predict(X_test,batch_size=16).reshape(205,1), label = 'Predict')
plt.legend()
plt.show()
```
